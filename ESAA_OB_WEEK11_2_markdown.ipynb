**개념 정리**

합성곱 신경망: 이미지의 국소적 부분을 계산함으로써 시간과 자원을 절약하여 이미지의 세밀한 부분까지 분석
- 이미지나 영상을 처리하는 데 유용
- 합성곱층: 이미지를 펼쳐서 분석했을 때 데이터의 공간적 구조를 무시하는 것을 방지

합성곱 신경망 구조:
- 입력층, 합성곱층, 풀링층, 완전연결층, 출력층
- 합성곱층과 풀링층을 거치면서 입력 이미지의 주요 특성 벡터를 추출 -> 완전연결층을 거치면서 1차원 벡터로 변환 -> 출력층에서 활성화 함수인 소프트맥스 함수를 사용하여 최종 결과 출력

입력층: 입력 이미지 데이터가 최초로 거치게 되는 계층
- 이미지: 높이, 너비, 채널의 값을 갖는 3차원 데이터
    - 채널: 이미지가 그레이스케일이면 1, 컬러이면 3

합성곱층: 입력 데이터의 특성을 추출하는 역할
- 커널/필터는 이미지의 모든 영역을 훑으면서 특성을 추출, 추출된 결과물이 특성 맵
- 커널은 3x3, 5x5 크기가 일반적, 스트라이드라는 지정된 간격에 따라 순차적으로 이동
- 컬러 이미지일 경우, RGB 각각에 서로 다른 가중치로 합성곱을 적용한 후 결과를 더함
- 요약:
    - 입력 데이터: W1 x H1 x D1
    - 하이퍼파라미터:
        - 필터 개수: K
        - 필터 크기: F
        - 스트라이드: S
        - 패딩: P
    - 출력 데이터:
        - W2 = (W1 - F + 2P) / S + 1
        - H2 = (H1 - F + 2P)/ S + 1
        - D2 = K

풀링층: 특성 맵의 차원을 다운 샘플링하여 연산량을 감소시키고, 주요한 특성 벡터 추출
- 최대 풀링: 대상 영역에서 최댓값을 추출
- 평균 풀링: 대상 영역에서 평균을 반환
    - 각 커널 값을 평균화시켜 가중치를 갖는 값의 특성이 희미해질 수 있음
- 요약:
    - 입력 데이터: W1 x H1 x D1
    - 하이퍼파라미터:
        - 필터 크기: F
        - 스트라이드: S
    - 출력 데이터:
        - W2 = (W1 - F) / S + 1
        - H2 = (H1 - F)/ S + 1
        - D2 = D1

완전연결층: 합성곱층과 풀링층을 거치면서 차원이 축소된 특성 맵이 전달
- 이미지는 3차원 벡터에서 1차원 벡터로 펼쳐짐

출력층: 소프트맥스 활성화 함수 사용, 입력받은 값을 0~1 사이의 값으로 출력
- 이미지가 각 레이블에 속할 확률 값이 출력, 가장 높은 확률 값을 갖는 레이블이 최종 값

1D 합성곱: 필터가 시간을 축으로 좌우로만 이동
- 입력(W)과 필터(k)에 대한 출력은 W가 됨

2D 합성곱: 필터가 방향 두 개로 움직이는 형태
- 입력(W, H)과 필터(k, k)에 대한 출력은 (W, H)가 됨, 출력 형태는 2D 행렬

3D 합성곱: 필터가 움직이는 방향이 3개
- 입력(W, H, L)과 필터(k, k, d)에 대한 출력은 (W, H, L)가 됨
    - d < L을 유지하는 것이 중요!!

3D 입력을 갖는 2D 합성곱
- 입력이 3D 형태임에도 출력 형태가 3D가 아닌 2D 행렬
- 필터에 대한 길이(L)가 입력 채널의 길이(L)와 같아야 하기 때문에 생성
- 입력(W, H, L)과 필터(k, k, L)에 대한 출력은 (W, H)가 됨
- 예시) LeNet-5, VGG

1x1 합성곱
- 입력은 3D 형태
- 입력(W, H, L)과 필터(1, 1, L)에 대한 출력은 (W, H)가 됨
- 예시) GoogLeNet
