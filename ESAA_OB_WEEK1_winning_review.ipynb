**주제 및 데이터**

주제: 2025 전력사용량 예측 AI 경진대회

https://dacon.io/competitions/official/236531/codeshare/12755?page=1&dtype=recent

데이터:
- train.csv: 85일 분량, 100개 건물들의 2024년 06월 01일부터 2024년 08월 24일까지의 데이터
    - 일시별 기온, 강수량, 풍속, 습도, 일조, 일사 정보, 전력사용량(kWh)
- building_info.csv: 100개 건물 정보
    - 건물 번호, 건물 유형, 연면적, 냉방 면적, 태양광 용량, ESS 저장 용량, PCS 용량
- test.csv: 100개 건물들의 2024년 08월 25일부터 2024년 08월 31일까지의 데이터
    - 일시별 기온, 강수량, 풍속, 습도의 예보 정보

**코드 리뷰**

(1) 전처리
- 한글 칼럼 영어로 변환
- building_info와 train, test를 건물번호 기준으로 merge
- 풍속과 습도가 0인 경우, 전과 후의 값을 바탕으로 선형 보간
- 시간 관련 파생 변수 생성
- 결측치 및 이상치 처리

(2) interpolation
- train의 일사량이 기간 전체에 0인 몇개의 건물을 xgb, lgbm으로 예측하여 보간
- test의 일조시간과 일조량을 예측하여 보간

(3) 모델링
- XGBoost와 LightGBM 활용한 앙상블
- optuna 하이퍼 파라미터 튜닝
- 세그멘트 별, 유형 별, 건물 별로 학습 예측

**차별점 및 배울점**

- interpolation을 여러 데이터에 대해 활용하였다.
- optuna 하이퍼 파라미터 튜닝을 적용하였다.
- 유형 별로 모델을 따로 학습해, 유형 고유의 패턴을 반영하였다.
