**주제 및 데이터**

주제: 주차 수요 예측 AI 경진대회

https://dacon.io/competitions/official/235745/overview/description

데이터:

- train.csv: 단지코드, 총세대수, 임대건물구분, 지역, 공급유형, 전용면적, 전용면적별세대수, 공가수, 신분, 임대료보증금, 임대료, 도보 10분거리 내 지하철역 수(환승노선 수 반영), 도보 10분거리 내 버스정류장 수,단지내주차면수, 등록차량수

- test.csv: 단지코드, 총세대수, 임대건물구분, 지역, 공급유형, 전용면적, 전용면적별세대수, 공가수, 신분, 임대료보증금, 임대료, 도보 10분거리 내 지하철역 수(환승노선 수 반영), 도보 10분거리 내 버스정류장 수, 단지내주차면수

- age_gender_info.csv: 지역 임대주택 나이별, 성별 인구 분포


**코드 리뷰**

(1) 패키지 및 데이터 로딩

(2) 데이터 전처리 및 시각화
- 칼럼명 변경
    - 도보 10분거리 내 지하철역 수(환승노선 수 반영)를 지하철로 변경
    - 도보 10분거리 내 버스정류장 수를 버스로 변경
- 데이터 타입 및 결측치 확인
    - 지하철, 버스 결측치 0으로 대체
    - 단지코드 결측치: 단지코드, 임대건물구분, 지역 등이 동일한 샘플의 값을 참고하여 대체
    - 실제 데이터를 기반으로 대체되지 못한 경우 0으로 대체 후 int로 변경
- 파생변수 생성
    - 등록차량수의 중앙값을 기준으로 지역을 5개 그룹으로 통합
    - 공급유형 '등록차량수-단지내주차면수' 차이를 기준으로 공급유형을 5개 그룹으로 통합
    - 전용면적을 5단위로 변환
- 1차원으로 데이터 병합

(3) 모형 학습, 예측 및 평가
- 단순 회귀
    - 피처와 타겟 간 상관 확인
    - 초기모형 학습 및 평가
        - Linear Regression
        - StandardScaler로 표준화
    - 회귀계수 확인을 통한 모형 검토
        - 정적 회귀계수 피처, 부적 회귀계수 피처 구분
    - permutation importance를 통한 피처 중요도 검토
- 다항회귀
    - 상호작용항 생성
    - PolynomialFeatures 클래스를 활용해 가장 적절한 2차식으로 표현

(4) 모형 개선
- 가중치 생성: Cook's D 활용
- Stratified KFold

**차별점 및 배울점**

- 정적 회귀계수 피처, 부적 회귀계수 피처로 구분했다.
- Permutation importance를 통해 피처 중요도를 검토한 점이 인상 깊었다.
- 가중치 생성을 위해 Cook's D를 활용했다.
