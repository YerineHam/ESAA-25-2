### 코드 리뷰

(1) 전처리
- 결측치 처리
    - 외부 데이터 활용, 동일한 지역/전용면적/총세대수를 가지는 단지를 조사하여 임대료와 임대보증금 값 처리
    - 선형회귀모형 이용
    - 지하철, 버스 결측치는 지역 평균으로 대체
- 데이터 가공 및 파생변수 생성
    - 자격유형 별 단지내주차면수와 등록차량수의 차이를 살펴본 뒤, 자격유형끼리 그룹화
    - 나이, 성별 데이터 활용, 남녀차이 변수/나이구분 변수 생성
    - 비슷한 지역끼리 그룹화하여 지역그룹 변수 생성
    - 분양세대수가 10보다 큰 단지에 대해 단지내주차면수를 조정
    - KNN 방법을 활용하여 변수 가공


(2) 모델링
- auto ML 툴인 Pycaret 사용
- 변수 선택: 상관관계, 피처 중요도 등을 활용하여 선택
- 여러 모델을 MAE 기준으로 비교한 뒤, catboost, huber, lasso 선택
    - 가장 높은 성능: Huber
    - 변수 선택에 효과적: Lasso
    - 범주형 변수 처리에 효과적: CatBoost

### 차별점 및 배울점

- 선형회귀모형으로 피팅을 해서 결측치를 처리했다.
- EDA를 전처리 과정 중간중간에 꾸준히 진행한 점이 좋았다.
- KNN 방법을 활용하여 변수를 가공하였다.
- 모델링할 때 Pycaret을 사용하였다.
