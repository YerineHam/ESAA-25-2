### 주제 및 데이터

주제: 시계열 기반 Superstore 판매량 예측

데이터: https://www.kaggle.com/datasets/rohitsahoo/sales-forecasting/data
- 글로벌한 superstore에 대한 4년치 자료
- 칼럼 별 활용 정보:
    - 주문 정보: Row ID, Order ID, Order Date, Ship Date, Ship Mode
    - 고객 정보: Customer ID, Customer Name, Segment
    - 지역 정보: Country, City, State, Postal Code, Region
    - 제품 정보: Product ID, Category, Sub-Category, Product Name
    - 매출 정보: Sales (Target 변수)

### 코드 리뷰

https://www.kaggle.com/code/ohseokkim/predicting-future-by-lstm-prophet-neural-prophet

(1) 전처리 및 EDA
- 결측치 처리: null 값이 존재하는 행 전체를 제거
- 고객 세그먼트 분석
    - Segment, Country, City, State 별 Sales 확인
    - Segment별, City별, State별 EDA 진행
    - Segment와 Category에 대한 Sales 확인
- Datetime 처리
    - Order Date, Shipe Date
    - Ship Date에서 Order Date를 뺀 것을 계산한 Processing Time 칼럼 추가
- Sales 트렌드 분석
    - Year별 가장 판매량이 높은 Category 및 Sub-Category 확인


(2) 모델링
- ARIMA, SARIMA 적용
    - auto_arima로 최적의 (p,d,q) 자동 탐색
    - seasonal 옵션 활성화
- LightGBM, Random Forest 적용
    - 과거 연도를 입력 변수로 사용해 패턴 학습
- LSTM 적용

-> 좋은 모델링이 아님!!!!!!!!!!!!!!!!!!!!

### 차별점 및 배울점


- 해당 코드처럼 여러 모델을 적용해보는 것도 좋지만, 각각의 목적이 확실해야 한다.
- 계절성이 거의 존재하지 않는 데이터에 seasonal ARIMA를 적용해서는 안된다.
- 연 단위로 데이터를 묶어버리면 관측치 수가 극히 적어져 좋은 결과가 나오지 않는다.
- 비선형 모델을 사용할 때는 시계열 특징 설계가 필요하며, lag를 반영하는 것이 좋다.
