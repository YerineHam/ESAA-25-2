## 주제 및 데이터

**주제:** Walmart Sales Analysis & Forecasting

https://www.kaggle.com/code/muhammadmudassir07/walmart-sales-analysis-forecasting-with-python#Time-Series-Forecasting

**데이터:**
- Walmart_Sales.csv
    - Store: 점포 번호
    - Date: 판매 주 시작일
    - Weekly_Sales: 주간 판매량
    - Holiday_Flag: 휴일 여부
    - Temperature: 해당 지역의 기온
    - Fuel_Price: 해당 지역의 기름값
    - CPI: 소비자 물가지수
    - Unemployment: 실업률


## 코드 리뷰

(1) 전처리
- 결측치 제거
- datetime 변환
- 중복 제거
- 이상치 처리 (IQR 방식 이용)

(2) EDA
- 시간에 따른 판매 추이
- 가게 별 판매 추이
- 공휴일 여부에 따른 판매
- 수치형 변수에 대한 상관관계

(3) 기술 통계량 확인
- 주요 변수 summary
- T-test를 활용하여 공휴일 여부에 따른 판매 확인

(4) 시계열 분석
- 하나의 가게를 선택, ARIMA/SARIMA 적용
    - SARIMAX 이용
- 다음 12주에 대한 판매 예측

(5) 모델
- 선형 회귀 적용
- 평가지표로 R2 적용

## 차별점 및 배울점

- 모델링 전에 T-test를 활용하여 판매를 확인했다.
- 시계열 분석을 위해 SARIMA를 적용하였다.
- SARIMAX 적용법에 대해 공부할 수 있었고, 미래 판매를 예측하는 과정을 확인할 수 있었다.
