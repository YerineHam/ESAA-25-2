### 주제 및 데이터

주제: 과거 Walmart 매출 데이터를 활용해 향후 매출을 예측

데이터: https://www.kaggle.com/competitions/walmart-recruiting-store-sales-forecasting/data
- 다양한 지역에 위치해 있는 45개의 Walmart에 대한 과거 매출 데이터
- stores.csv: 각 매장의 type와 size
- train.csv: 2010-02-05부터 2012-11-01까지의 데이터
    - Store: 매장 번호
    - Dept: 부서 번호
    - Date: 주 단위 날짜
    - Weekly_Sales: 해당 매장/부서의 주간 매출
    - IsHoliday: 공휴일이 포함된 주인지 여부
- test.csv: Weekly_Sales 제외 train과 동일
- features.csv: 특정 날짜의 매장, 부서, 지역활동과 관련된 추가 정보
    - Store: 매장 번호
    - Date: 주 단위 날짜
    - Temperature: 해당 지역의 평균 기온
    - Fuel_Price: 해당 지역의 연료 가격
    - Markdown1 ~ Markdown5: Walmart에서 진행한 프로모션 할인 관련 데이터
    - CPI: 소비자 물가 지수
    - Unemployment: 실업률
    - IsHoliday: 공휴일이 포함된 주인지 여부

### 코드 리뷰

https://www.kaggle.com/code/sayamkumar/walmart-store-sales-forecasting-using-arima/notebook

(1) 전처리 및 EDA
- 전처리
    - Weekly_Sales가 0보다 큰 경우만 유지
    - Date를 DatetimeIndex로 변환 후 정렬
    - 데이터를 주간 단위별 총매출 데이터로 단순화
    - 이동평균 기반 변수 생성
        - SMA-7, SMA-30: 단순 이동평균
        - EWMA-7, EWMA-30: 지수 가중 이동평균
- EDA
    - 평균 주간 매출 및 최대 매출 확인
        - 공휴일의 영향이 강하고 변동성이 큼
    - 이동평균 기반 추세 확인





(2) 모델링
- Simple Exponential Smoothing 적용
    - trend(추세), seasonal(계절성) 미적용
    - Baseline으로 이용
- Holt Winters Model 적용
    - trend, seasonal 반영
- ETS 모델 적용
    - error, trend, seasonal 모두 반영
    - Holt Winters Model과 유사한 예측 패턴을 보임
- Holt 적용
    - 계절성을 제거하고, 장기 추세만 반영
    - 공휴일 효과를 반영하는 데 실패함
- Stationary Test
    - ADF test / KPSS test
    - 정상성 만족 확인
- AR 모델 적용
    - AR 차수 선택: Seasonal AR(1)
    - 추세 포함, 외생변수 IsHoliday
- MA(q) 모델 적용
    - ARIMA(0,0,3): MA(3), 현재 매출이 과거 3주간의 오차에 의해 설명
    - 단기 오차 구조는 설명하지만, 연간 계절성을 설명하기에는 부족
- ARIMA 적용
    - ARIMA(2,0,1): AR(2) - 과거 2주 매출 영향 / MA(1) - 오차 1항
    - Auto ARIMA
        - p,d,q 자동 탐색
        - 외생변수 IsHoliday 포함
        - ARIMA(2,0,0) 선택
- SARIMA 적용
    - SARIMA(2,0,1)(1,0,1,52)
    - 매출의 연간 계절 구조를 가장 잘 설명


### 차별점 및 배울점


- 다양한 유형의 시계열 예측 모델을 적용해봄으로써 잘 알지 못했던 모델의 모델링 과정까지 경험해볼 수 있었다.
- 처음부터 ARIMA와 SARIMA를 적용한 것이 아니라, 단계적으로 모델의 복잡도를 올린 점이 좋았다.
- Auto ARIMA는 최적 모델을 찾는 것이 아니라, 비계절 ARIMA 중에서 최적을 찾는 것이다.
