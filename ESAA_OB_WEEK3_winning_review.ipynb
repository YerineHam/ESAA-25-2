**주제 및 데이터**

주제: 제주도 도로 교통량 예측 AI 경진대회

https://dacon.io/competitions/official/235985/overview/description

데이터:
- train.csv: 2022년 8월 이전 데이터 (단, 날짜가 연속적이지 않음)
    - id: 샘플 별 고유 id
    - 날짜, 시간, 교통 및 도로구간 등 정보
    - target: 도로의 차량 평균 속도(km)
- test.csv: 2022년 8월 데이터만 존재 (단, 날짜가 연속적이지 않음)
    - id: 샘플 별 고유 id
    - 날짜, 시간, 교통 및 도로구간 등 정보
- data_info.csv: 데이터의 각 column 별 추가 설명



**코드 리뷰**

(1) 전처리 및 EDA
- csv to parquet: 메모리에 효율적인 데이터 유형을 사용, 빠른 작업 가능
- 파생 변수 생성
    - 주요 항목 별 target의 train 데이터셋 평균과 표준편차
    - 시간, 제한속도 별 평균
    - 시간 별 target 평균에 대한 시간, 제한속도 별 target 평균의 비율
- 시간 cos/sin 변환 추가
- 공휴일 변수 생성
- 주요 변수 별 target의 분포 확인: 박스플롯 활용

(2) 모델링
- 모델 생성 함수 활용
    - feature_fraction, bagging_fraction 등의 파라미터 적용
- 데이터 분할 파라미터 및 모델 파라미터 조합으로 앙상블
    - LGBM 모델의 학습률과 seed를 다양하게 조합
    - 데이터를 99:1, 95:5로 랜덤 분할 (StratifiedShuffleSplit)
- 변수 중요도 확인
    - split 중요도, gain 중요도

**차별점 및 배울점**

- 시간 데이터에 대해 sin/cos 변환을 적용한 점이 인상 깊었다.
- 모델 생성 함수를 정의할 때 처음 보는 파라미터를 많이 활용하였다.
- 모델링 과정에서 데이터를 랜덤으로 분할하였다.
- 변수 중요도가 importance type에 따라 나뉠 수 있다는 것을 알게 되었다.
