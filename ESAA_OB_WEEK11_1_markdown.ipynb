### 4.1 인공 신경망의 한계와 딥러닝 출현

퍼셉트론: 다수의 신호를 입력으로 받아 하나의 신호를 출력
- AND 게이트: 모든 입력이 '1'일 때 작동
- OR 게이트: 입력에서 둘 중 하나만 '1'이거나 둘 다 '1'일 때 작동
    - 모두 '0'인 경우만 제외
- XOR 게이트: 배타적 논리합, 입력 중 한 개만 '1'일 때 작동
    - 데이터가 비선형적으로 분리되기 때문에 제대로 된 분류가 어려움
    - 단층 퍼셉트론에서 XOR에 대한 학습이 불가능
    - 이를 해결하기 위해 다층 퍼셉트론 고안

딥러닝: 입력층과 출력층 사이에 은닉층이 여러 개 있는 신경망

### 4.2 딥러닝 구조

딥러닝 구성 요소
- 층: 입력층, 은닉층 출력층
- 가중치: 노드와 노드 간 연결 강도
    - 입력 값이 연산 결과에 미치는 영향력을 조절
- 바이어스: 가중합에 더해 주는 상수
- 가중합(=전달 함수): 가중치와 신호의 곱을 합한 것
- 함수: 활성화 함수, 손실 함수

활성화 함수: 전달 함수에서 전달 받은 값을 출력할 때 일정 기준에 따라 출력 값을 변화시키는 비선형 함수
- 시그모이드 함수: 선형 함수의 결과를 0~1 사이에서 비선형 형태로 변형
    - 기울기 소멸 문제 발생, 딥러닝 모델에서는 잘 사용 X
- 하이퍼볼릭 탄젠트 함수: 선형 함수의 결과를 -1~1 사이에서 비선형 형태로 변형
    - 시그모이드에서 결괏값의 평균이 0이 아닌 양수로 편향된 문제를 해결
- 렐루 함수: 입력(x)이 음수일 때는 0을 출력, 양수일 때는 x를 출력
    - 학습 속도가 빠름, 기울기 소멸 문제 발생 X
    - 일반적으로 은닉층에서 사용
    - 리키 렐루: 음수 값을 입력 받으면 항상 0을 출력하는 문제 해결
- 리키 렐루 함수: 입력 값이 음수이면 0이 아닌 매우 작은 수를 반환
    - 입력 값이 수렴하는 구간이 제거
- 소프트맥스 함수: 입력 값을 0~1 사이에 출력되도록 정규화하여 출력 값들의 총합이 항상 1이 되도록 함

손실 함수: 학습을 통해 얻은 데이터의 추정치가 실제 데이터와 얼마나 차이가 나는지 평가하는 지표
- 평균 제곱 오차: 실제 값과 예측 값의 차이를 제곱하여 평균을 낸 것
- 크로스 엔트로피 오차: 분류 문제에서 원-핫 인코딩을 했을 때 사용할 수 있는 오차 계산법
    - 두 개의 확률 분포 차이를 이용하기 때문에 시그모이드의 영향을 덜 받음


딥러닝 학습
- 순전파: 네트워크에 훈련 데이터가 들어올 때 발생, 데이터를 기반으로 예측 값을 계산하기 위해 전체 신경망을 교차
    - 모든 뉴런이 이전 층의 뉴런에서 수신한 정보에 변환을 적용하여 다음 층의 뉴런으로 전송
- 역전파: 손실이 계산되면 그 정보를 역으로 전파
    - 예측 값과 실제 값 차이를 각 뉴런의 가중치로 미분한 후 기존 가중치 값에서 뺌

딥러닝의 문제점과 해결 방안
- 과적합 문제 발생: 훈련 데이터를 과하게 학습
    - 해결: 드롭아웃, 학습 과정 중 임의로 일부 노드들을 학습에서 제외
- 기울기 소멸 문제 발생: 출력층에서 은닉층으로 전달되는 오차가 크게 줄어들어 학습이 되지 않음
    - 해결: 시그모이드/하이퍼볼릭 탄젠트 대신 렐루 활성화 함수 사용
- 성능이 나빠지는 문제 발생
    - 해결: 확률적 경사 하강법과 미니 배치 경사 하강법 사용

경사 하강법
- 배치 경사 하강법: 전체 데이터셋에 대한 오류를 구한 후 기울기를 한 번만 계산하여 모델의 파라미터를 업데이트
    - 전체 훈련 데이터셋에 대해 가중치를 편미분하는 방법
    - 한 스텝에 모든 훈련 데이터셋을 사용, 학습이 오래 걸리는 단점
- 확률적 경사 하강법: 임의로 선택한 데이터에 대해 기울기를 계산
- 미니 배치 경사 하강법: 전체 데이터셋을 미니 배치 여러 개로 나누고, 각각의 기울기를 구한 후 평균 기울기를 이용하여 모델을 업데이트
    - 확률적 경사 하강법보다 안정적

옵티마이저: 학습 속도와 운동량을 조정
- 속도 조정: 아다그라드, 아다델타, 알엠에스프롭
- 운동량 조정: 모멘텀, 네스테로프 모멘텀
- 속도와 운동량에 대한 혼용: 아담

딥러닝을 사용할 때 이점
- 특성 추출: 데이터 별로 특징 찾아내고, 그것을 토대로 데이터를 벡터로 변환
    - 딥러닝은 이를 알고리즘에 통합
- 빅데이터의 효율적 활용

### 4.3 딥러닝 알고리즘

심층 신경망: 입력층과 출력층 사이에 다수의 은닉층을 포함하는 인공 신경망
- 비선형적 관계 학습 가능
- 학습을 위한 연산량이 많고 기울기 소멸 문제 발생 가능

합성곱 신경망: 합성곱층과 풀링층을 포함하는 이미지 처리 성능이 좋은 인공 신경망 알고리즘
- LeNet-5, AlexNet, VGG, GoogLeNet, ResNet 등
- 각 층의 입출력 형상을 유지, 이미지의 공간 정보를 유지하면서 인접 이미지와 차이가 있는 특징을 효과적으로 인식
- 복수 필터로 이미지의 특징을 추출 및 학습
- 추출한 이미지의 특징을 모으고 강화하는 풀링층이 있음

순환 신경망: 시계열 데이터 같은 시간 흐름에 따라 변화하는 데이터를 학습하기 위함
- 시간에 따라 내용이 변하므로 데이터는 동적, 길이는 가변적
- 기울기 소멸 문제 발생 -> LSTM 도입

제한된 볼츠만 머신: 가시층과 은닉층으로 구성, 가시층은 은닉층과만 연결
- 차원 감소, 분류, 선형 회귀 분석, 협업 필터링, 특성 값 학습, 주제 모델링에 사용
- 기울기 소멸 문제를 해결하기 위해 사전 학습 용도로 활용 가능
- 심층 신뢰 신경망의 요소로 활용

심층 신뢰 신경망: 제한된 볼츠만 머신을 여러 층으로 쌓은 형태로 연결
- 학습 절차: 가시층과 은닉층1에 제한된 볼츠만 머신을 사전 훈련 -> 첫 번째 층 입력 데이터와 파라미터를 고정하여 두 번째 층 제한된 볼츠만 머신을 사전 훈련 -> 원하는 층 개수만큼 제한된 볼츠만 머신을 쌓아 올려 전체 DBN 완성

```python
class Net(torch.nn.Module):
    def __init__(self, n_feature, n_hidden, n_output):
        super(Net, self).__init__()
        self.hidden=torch.nn.Linear(n_feature, n_hidden)
        self.relu=torch.nn.ReLu(inplace=True)
        self.out=torch.nn.Linear(n_hidden, n_output)
        self.softmax=torch.nn.Softmax(dim=n_output)
    def forward(self, x):
        x=self.hidden(x)
        x=self.relu(x)
        x=self.out(x)
        x=self.softmax(x)
        return x
```
